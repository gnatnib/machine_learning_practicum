# **I. Import Library**

```python
%matplotlib inline
#%matplotlib inline sets the backend o matplotlib to the 'inline' backend

import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
from sklearn.cluster import KMeans
```

# **II. Load Dataset**

```python
#buat dataset
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

#Mengambil data dari fitur dataset iris
plt.plot()
plt.scatter(features[0], features[1])
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

#**III. Menentukan nilai yang tepat untuk cluster**

```python
X = np.array(list(zip(features[0], features[1]))).reshape(len(features[1]),2)
SSE = []
K = range(1,10)
for k in range (1,10):
  kmeanModel = KMeans(n_clusters = k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)
```

#**IV. Plot nilai K**

```python
#Plotting the elbow
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

# **V. Menentukan nilai K berdasar elbow graph tersebut**

```python
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X)

LABEL_COLOR_MAP = {0: 'b', 1: 'g', 2: 'r'}
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]

plt.scatter(features[0], features[1], c=label_color)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('K=3')
plt.show()
```

```python
figure,ax=plt.subplots(2,3)
K=range(1,7)

for k in K:
  if(k<4):
    row = 0
    column = k-1
  else:
    row=1
    column = k-4

  kmeanModel = KMeans(n_clusters=k).fit(X)
  y_pred = kmeanModel.fit_predict(X)
  ax[row][column].scatter(features[0], features[1], c=y_pred, cmap = 'winter')

plt.show()
```

```python
#buat dataset
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

#Mengambil data dari fitur dataset iris
plt.plot()
plt.scatter(features[1], features[2])
plt.xlabel('Sepal width')
plt.ylabel('Petal length')
plt.show()
```

```python
X = np.array(list(zip(features[1], features[2]))).reshape(len(features[2]),2)
SSE = []
K = range(1,10)
for k in range (1,10):
  kmeanModel = KMeans(n_clusters = k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)
```

```python
#Plotting the elbow
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X)

LABEL_COLOR_MAP = {0: 'b', 1: 'g', 2: 'r'}
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]

plt.scatter(features[1], features[2], c=label_color)
plt.xlabel('Sepal width')
plt.ylabel('Petal length')
plt.title('K=3')
plt.show()
```

```python
cmaps =['coolwarm', 'viridis', 'plasma', 'magma']
```

```python
for cmap in cmaps:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    fig.suptitle(f'Clustering dengan colormap {cmap}')

    for k in range(1, 7):
        row = (k-1) // 3
        col = (k-1) % 3

        kmeanModel = KMeans(n_clusters=k).fit(X)
        y_pred = kmeanModel.fit_predict(X)

        axes[row, col].scatter(features[1], features[2], c=y_pred, cmap=cmap)
        axes[row, col].set_title(f'K={k}')
        axes[row, col].set_xlabel('Sepal Width')
        axes[row, col].set_ylabel('Petal Length')

    plt.tight_layout()
    plt.show()
```
